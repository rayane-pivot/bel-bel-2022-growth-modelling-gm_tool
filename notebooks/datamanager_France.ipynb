```python
import pandas as pd
import numpy as np
import datetime
```

```python
dict_path = {
    'ALL' : 'data/France/Growth Modelling - FR - 2019-2021 - Panel Data (IRI) - Total GMS.xlsx',
    'Concepts HMSM' : 'data/France/Growth Modelling - FR - 2019-2021 - Panel Data (IRI) - Concepts HMSM.xlsx',
    'Ecommerce' : 'data/France/Growth Modelling - FR - 2019-2021 - Panel Data (IRI) - Ecommerce.xlsx',
    'Hard Discount' : 'data/France/Growth Modelling - FR - 2019-2021 - Panel Data (IRI) - Hard Discount.xlsx',
    'Proxi' : 'data/France/Growth Modelling - FR - 2019-2021 - Panel Data (IRI) - Proxi.xlsx'
}
```

```python
df_concat = pd.DataFrame()
for distrib, path in dict_path.items():
    df = pd.read_excel(path, header=[7, 8], sheet_name=[str(i) for i in range(1, 16)])
    df = pd.concat(df.values(), axis=0)

    df.columns = df.columns.droplevel(0)
    cols=[
        "Date",
        "Product",
        "Category",
        "Sub Category",
        "Brand",
        "Sales in value",
        "Sales value with promo",
        "Sales in volume",
        "Sales volume with promo",
        "Price with promo",
        "Price without promo",
        "Price per volume",
        "DV Cum (EAN)"
        ]
        
    df = df.rename(columns={df.columns[i]:cols[i] for i, col in enumerate(cols)})
    df.Date = df.Date.apply(lambda x:datetime.datetime.strptime(x[-10:], '%d-%m-%Y').strftime('%Y-%m-%d'))
    df['Distribution Route'] = distrib
    df_concat = pd.concat([df_concat, df])
```
